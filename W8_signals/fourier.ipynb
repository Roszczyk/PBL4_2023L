```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs=1
N=100

t=np.arange(N)
s=np.sin(0.15*2*np.pi*t)
S=np.fft.fftshift(np.fft.fft(s))
f=np.arange(Fs/-2, Fs/2, Fs/N)

S_mag=np.abs(S)
S_phase=np.angle(S)
plt.figure(0)
plt.plot(f,S_mag, ".-")
plt.plot(f,S_phase, ".-")
plt.legend(["Widmo amplitudowe", "Widmo fazowe"], loc=1)
plt.xlabel("Częstotliwość względem Fs")
plt.grid()
plt.show()
```

```python
c=np.cos(0.15*2*np.pi*t)
simag=1j*s
sygn=c+simag

Sygn=np.fft.fftshift(np.fft.fft(sygn))
f=np.arange(Fs/-2, Fs/2, Fs/N)

Sygn_mag=np.abs(Sygn)
Sygn_phase=np.angle(Sygn)
plt.figure(0)
plt.plot(f,Sygn_mag, ".-")
plt.plot(f,Sygn_phase, ".-")
plt.legend(["Widmo amplitudowe", "Widmo fazowe"], loc=1)
plt.xlabel("Częstotliwość względem Fs")
plt.grid()
plt.show()
```
