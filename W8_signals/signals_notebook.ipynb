```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.arange(100)
s=np.sin(0.99*2*np.pi*t)

plt.figure(0)
plt.plot(s, ".-")

plt.show()
```

```python
Fs=50000
Ts=1.0/Fs
f0=1500
T=0.01

n=np.arange(Fs*T)
t=n/Fs
s=np.sin(f0/Fs*2*np.pi*n)

plt.figure(0)
plt.plot(t,s, ".-")
plt.xlabel("Czas[s]")
plt.show()
```

```python
f1=200
f3=600
f5=1000
f7=1400

a1=1
a3=1/3
a5=1/5
a7=1/7

s1=a1*np.sin(f1/Fs*2*np.pi*n)
s3=a3*np.sin(f3/Fs*2*np.pi*n)
s5=a5*np.sin(f5/Fs*2*np.pi*n)
s7=a7*np.sin(f7/Fs*2*np.pi*n)

sygn=s1+s3+s5+s7

plt.figure(0)
plt.plot(t,sygn, ".-")
plt.xlabel("Czas[s]")
plt.show()
```
